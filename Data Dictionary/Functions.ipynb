```
# This file is for miscellaneous functions that might be used across multiple notebooks primarily used in the DataDictionary library


"""
    Lists the full filepath of every file that can be found in the given directory. The purpose of this function is to easily load an entire folder of datasets in ADLS with ease and with flexibile configuration options.

    Parameters:
    directory (string): The directory you want to search
    search_dirs (bool, optional): Whether or not you want to search further directories found withint he given directory. Defaults to True
    filetype (string, optional): The file extension you are searching for. Defaults to .csv
    full_path (bool, optional): Whether or not you want the full path of each file. Defaults to True
        ex:
            True -> dbfs:/mnt/team-juniper/Landing Zone/Our World in Data - Energy/Energy Consumption by Source & Country/energy-consumption-by-source-and-country.csv
            False -> energy-consumption-by-source-and-country.csv

    Returns:
    list: A list of files/filepaths. The files/filepaths are expressed as a string
"""
def list_files(directory, search_dirs=True, filetype='.csv', full_path=True):
    files = dbutils.fs.ls(directory)
    file_list = []
    for file in files:
        if filetype in file.name:
            if full_path:
                file_list.append(file.path)
            else:
                file_list.append(file.name)
        elif file.isDir():
            if search_dirs:
                file_list.extend(list_files(file.path, search_dirs, filetype, full_path))

    return file_list
```

```
"""
    Searches each path in the list of given paths and extracts filenames of the given extension type. Very useful in conjunction with list_files() to extract the file names of both regular .csv files and spark partition files. This helps a lot with keeping consistent naming in the pipeline.

    Parameters:
        paths (list): A list of strings that are the absolute filepaths of the files to extract the names of
        filetype (string, optional): The file extension to look for. Defaults to .csv

    Returns:
    list: A list of file names extracted from the given path list
"""
def getFileNames(paths, filetype='.csv'):
    return [file.split(filetype)[0].split('/')[-1] for file in paths]
```

```
"""
    Returns true if the dataframe has a column with the given name
"""
def has_column(df, col_name):
    return col_name in df.columns
```

```
"""
    Displays all dataframes in a given list
"""
def displayAll(dfs):
    for df in dfs:
        display(df)
```
